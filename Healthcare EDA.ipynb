```python
df['sex'].replace({0:'FEMALE', 1:'MALE'}, inplace=True)
df['cp'].replace({0:'Asymptomatic', 1:'Typical Angina', 2:'Atypical Angina', 3:'Non-anginal Pain'}, inplace=True)
df['fbs'].replace({0:'False', 1:'True'}, inplace=True)
df['restecg'].replace({0:'Hypertrophy', 1:'Normal', 2:'ST-T Wave Abnormality'}, inplace=True)
df['exng'].replace({0:'NO', 1:'YES'}, inplace=True)
df['slp'].replace({0:'Downsloping', 1:'Flat', 2:'Upsloping'}, inplace=True)
df['thall'].replace({1:'Fixed Defect', 2:'Normal', 3:'Reversable Defect'}, inplace=True)
df['exng'].replace({0:'NO', 1:'YES'}, inplace=True)
df['output'].replace({0:'Less Chance', 1:'More Chance'}, inplace=True)
```

```python
df.head()
```

```python
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exng', 'slp', 'caa', 'thall', 'output']
num_cols = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']
```

##  Univariate Analysis

### Categorical Analysis

```python

palette_color = sns.color_palette('Accent')

for col in cat_cols:

    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    # Countplot
    ax = sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index,
        ax=axs[0],
        hue=col,
        palette='Accent',
        legend=False
    )

    for container in ax.containers:
        ax.bar_label(container, fontsize=11)

    axs[0].set_xlabel(col, fontsize=14)
    axs[0].set_ylabel('Count', fontsize=14)
    axs[0].set_title(f'Count of {col}', fontsize=16)

    # Pie Plot
    df[col].value_counts().plot.pie(
        ax=axs[1],
        ylabel='',
        autopct='%.2f%%',
        colors=[palette_color[i] for i in range(df[col].nunique())]
    )

    axs[1].set_title(f'Ratio of {col}', fontsize=16)

    plt.tight_layout()
    plt.show()
```

# Insights of Categorical Analysis

## Sex

- 207 patients are male, while 96 are female.
- 68.32% of the patients are male and 31.68% are female.
- The number of male patients in the dataset is 2.15 times the number of female patients.

---

## CP (Chest Pain)

- Almost half of the patients in the dataset have **asymptomatic chest pain**.
  - A total of **143 patients** experienced this type of pain, representing **47.19%** of all patients.
- Just over a quarter of the patients have **atypical angina**.
  - A total of **87 patients** experienced this type of chest pain, accounting for **28.71%** of the dataset.
- The remaining chest pain types are less common:
  - **Typical angina:** 50 patients (**16.50%**)
  - **Non-anginal pain:** 23 patients (**7.59%**)

---

## FBS (Fasting Blood Sugar)

- If a patient's fasting blood sugar is lower than **120 mg/dl**, the patient is considered non-diabetic (**0 / False**).
- If a patient's fasting blood sugar is higher than **120 mg/dl**, the patient is considered diabetic (**1 / True**).
- **45 patients (14.85%)** in the dataset have diabetes.
- **258 patients (85.15%)** do not have diabetes.

---

## RestECG (Resting Electrocardiographic Results)

- The first observation is that the two main categories are almost equally distributed and together account for nearly the entire dataset.
- The third category is negligible in comparison.
- **152 patients (50.17%)** have normal resting electrocardiographic results.
- **147 patients (48.51%)** have hypertrophy.
- Only **4 patients (1.32%)** have an **ST-T wave abnormality**.

---

## Exang (Exercise-Induced Angina)

- The variable **exang** indicates whether exercise-induced angina occurs:
  - **1 = Yes**
  - **0 = No**
- Patients without exercise-induced angina outnumber those with it by approximately **2.06 times**.
- **99 patients (32.67%)** experience exercise-induced angina.
- **204 patients (67.33%)** do not experience exercise-induced angina.

---

## SLP (Slope of the Peak Exercise ST Segment)

- The least common category is **downsloping ST segment**, observed in **21 patients (6.93%)**.
- The other two categories are almost equally distributed:
  - **Flat ST segment:** 142 patients (**46.86%**)
  - **Upsloping ST segment:** 140 patients (**46.20%**)

---

## CAA (Number of Major Vessels)

- This variable represents the number of major vessels colored by fluoroscopy.
- More than half of the patients (**175 patients, 57.76%**) have **0 major vessels** detected.
- The second most common value is **1 major vessel**, observed in **65 patients (21.45%)**.

---

## Thall (Thallium Stress Test)

- The Thallium stress test is an imaging procedure used to evaluate blood flow to the heart muscle and assess the presence of coronary artery disease.
- Although the variable description lists three categories, the dataset contains four values: **0, 1, 2, and 3**.
- Based on further investigation, the value **0** appears to represent missing or null observations.
- **166 patients (54.79%)** have a **normal** Thallium stress test result.
- **111 patients (36.61%)** have a **reversible defect**.
- **18 patients (5.94%)** have a **fixed defect**.

---

## Output (Heart Disease Risk)

- More than half of the patients are classified as having a risk of heart disease.
- **165 patients (54.46%)** are categorized as having a higher risk.
- **138 patients (45.54%)** are categorized as having a lower risk.


## Numerical Analysis

```python

palette_color = sns.color_palette('Set2')

for idx, col in enumerate(df[num_cols]):
    
    plt.figure(figsize=(20, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    df[col].hist(
        bins=50,
        alpha=0.6,
        color=palette_color[idx]
    )

    plt.axvline(np.mean(df[col]), ls='--', c='r', label='Mean')
    plt.axvline(np.median(df[col]), ls=':', c='g', label='Median')

    plt.ylabel("Counts")
    plt.title(f"{col} Histogram Distribution")
    plt.legend()

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color=palette_color[idx])

    plt.title(f"{col} Boxplot")

    plt.tight_layout()
    plt.show()
```

# Insights of Numerical vs Target Analysis

## Age vs Target

- Observations suggest that, contrary to common assumptions, the risk of heart disease appears to decrease with age in this dataset.
- While it is generally believed that older individuals are more likely to experience heart disease, the age distribution indicates a different pattern.
- The correlation between **Age** and **Output** is **-0.23**, indicating a **low negative correlation**.

---

## Trtbps (Resting Blood Pressure) vs Target

- The distributions of patients with and without heart disease are very similar across resting blood pressure values.
- As a result, resting blood pressure alone does not appear to be a strong predictor of heart disease risk.
- The correlation between **Trtbps** and **Output** is **-0.14**, indicating a **low negative correlation**.

---

## Chol (Cholesterol) vs Target

- Cholesterol values between **200 and 250** appear to be a critical range for heart disease risk.
- At lower cholesterol levels, the distributions of patients with and without heart disease are relatively similar.
- Between **180 and 250**, the probability of heart disease increases noticeably.
- Beyond a cholesterol value of **250**, distinguishing between the two groups becomes more difficult.
- The correlation between **Chol** and **Output** is **-0.085**, indicating a **very weak negative correlation**.

---

## Thalachh (Maximum Heart Rate Achieved) vs Target

- A clear trend is visible: as the maximum heart rate achieved increases, the likelihood of heart disease also increases.
- Up to a value of approximately **150**, both target groups increase at a similar rate.
- Beyond **150**, the number of patients with a lower risk decreases, while the number of patients with a higher risk increases significantly.
- The separation between the two distributions suggests a meaningful relationship between the variables.
- The correlation between **Thalachh** and **Output** is **0.42**, indicating a **moderately strong positive correlation**.

---

## Oldpeak vs Target

- When the **Oldpeak** value lies between **0 and 1.5**, the probability of heart disease increases substantially.
- The range of **0 to 1.5** can be considered a critical threshold.
- The distinction between the distributions of the two target groups suggests a notable relationship between the variables.
- The correlation between **Oldpeak** and **Output** is **-0.43**, indicating a **moderately strong negative correlation**.
